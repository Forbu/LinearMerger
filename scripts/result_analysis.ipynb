```python
import numpy as np
import matplotlib.pyplot as plt

x_axis_values = np.linspace(0, 1, 100)

results_simple = np.load("../results/merge_simple.npy")
results_linearmerge = np.load("../results/merge_linearmerge.npy")

plt.plot(x_axis_values, results_simple, label="Simple")
plt.plot(x_axis_values, results_linearmerge, label="LinearMerge")

# label the x axis
plt.xlabel("Interpolation value")
plt.ylabel("Test loss")

plt.grid()
plt.legend()
plt.show()


```
